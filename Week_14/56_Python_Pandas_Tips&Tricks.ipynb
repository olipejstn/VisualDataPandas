Persiapan Data Frame

```python
data = {'nama':['bayu', 'indra', 'devi', 'agni'],
        'jenis_kelamin' : ['L', 'L', 'P', 'L'], 
        'usia' : [23, 21, 22, 25]}

df = pd.DataFrame(data)
df
```

Mengeluarkan kolom dari Data Frame

```python
usia = df.pop('usia').to_frame()
```
